```python
from osgeo import gdal, ogr

tiff_filename = r'/content/drive/MyDrive/Results/Ahmd_VV.tif'
png_filename = r'/content/drive/MyDrive/Results/Ahmd_VV.png'

dataset = gdal.Open(tiff_filename)
print(dataset.RasterCount)

band_1 = dataset.GetRasterBand(1)
b1 = band_1.ReadAsArray()
print(b1)

options_list = ['-ot Byte','-of PNG','-b 1','-scale']


options_string = " ".join(options_list)

gdal.Translate(png_filename, tiff_filename, options=options_string)
```
